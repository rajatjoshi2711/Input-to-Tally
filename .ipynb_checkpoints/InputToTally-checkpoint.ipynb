# Queries

### Read

```python
conn = mod.connect("Url='http://localhost:9000'")
cur = conn.cursor()

cur.execute("SELECT Name, Address FROM Company")
rs = cur.fetchall()
print('Companies:')
for row in rs:
    print(row)

cur.execute("SELECT Name FROM Groups WHERE Name like '%Demo%'")
rs = cur.fetchall()
print('\nGroups:')
for row in rs:
    print(row)
    
cur.execute("SELECT Name, Bankaccholdername, Narration, StateName, ClosingBalance FROM Ledger")
rs = cur.fetchall()
print('\nLedgers:')
for row in rs:
    print(row)

cur.execute("SELECT Name, CreditTotals, ClosingBalance FROM VoucherCostCategory")
rs = cur.fetchall()
print('\nLedgers:')
for row in rs:
    print(row)
```

### Insert

```python
cur.execute("INSERT INTO Company(Name, Address) VALUES ('myAwesomeCompany', 'Silicon Valley')")
```

```python
cur.execute("INSERT INTO Ledger(Name, Country) VALUES ('myAwesomeLedger', 'India')")
```

### Update

```python
cur.execute("UPDATE Groups SET Name = 'New_Demo_Group' WHERE Name = 'Demo_Group'")
```

### List of all tables and columns inside them

```python
#List of all tables
cur.execute("SELECT TableName FROM sys_tables")
rs = cur.fetchall()
print('\nAll Tables:')
for row in rs:
    print(str(row[0]))
```

```python
#List of all tables along with columns in them
cur.execute("SELECT TableName FROM sys_tables")
rs = cur.fetchall()
print('\nAll Tables:')
for row in rs:
    print('\n')
    print(str(row[0]))
    cur.execute("Select ColumnName FROM sys_tablecolumns WHERE TableName = '" + str(row[0]) + "'")
    rs2 = cur.fetchall()
    print(">> Columns in " + str(row[0]) + ":")
    for row2 in rs2:
        print(str(row2[0]))
```

```python
# Detailed description of columns in a table
cur.execute("SELECT * FROM sys_tablecolumns WHERE TableName = 'Ledger' and ColumnName = 'ClosingBalance'")
rs = cur.fetchall()
print('\nLedger columns:')
for row in rs:
    print(row)
```
